```python
import random
from IPython.display import clear_output as co

class Deck:
    def __init__(self):
        self.suits = ['Diamond', 'Heart', 'Club', 'Spade']
        self.ranks = [2, 3, 4, 5, 6, 7, 8, 9, 10, 10, 10, 10, 11]
        self.cards = [(r, s) for s in self.suits for r in self.ranks]
    
    def shuffle(self):
        print('Dealer is shuffling... \n')
        random.shuffle(self.cards)
    
class Player:
    def __init__(self):
        self.hand = []
        
    @classmethod
    def get_total(self, cls):
        total = sum([card[0] for card in cls.hand])
        if total > 21:
            print(cls.bust())
        return total
    

    def bust(self):
        return f'{self.__class__.__name__} Bust!'
    
    
    

class Dealer(Player):
    def deal(self, a_deck, cls):
        # if human has more than 2 cards
        if isinstance(cls, Human):
            if len(cls.hand) >= 2:
                # perform a 'hit'
                cls.hand.append(a_deck.cards.pop())
            else:
                # otherwise, the player needs to be dealt all of their cards
                for _ in range(2):
                    cls.hand.append(a_deck.cards.pop())
        # dealer deals one card to themselves at a time
        if isinstance(cls, Dealer):
            self.hand.append(a_deck.cards.pop())

class Human(Player):
    pass

class Game:
    
    @classmethod
    def run(self):
        print("Welcome to PyJack!")
        deck = Deck()
        dealer = Dealer()
        human = Human()
        
        def clear_hand():
            human.hand = []
            dealer.hand =[]
            dealer.deal(deck, human)
            dealer.deal(deck, dealer)
        
        deck.shuffle()
        
#         print(deck.cards)
        
        dealer.deal(deck, human)
        
#         print(len(deck.cards))
        print(human.hand)
        print(f'Total:  {Player.get_total(human)}')
        
        dealer.deal(deck, dealer)
        
#         print(len(deck.cards))
        print(dealer.hand)
        
        done = False
        
        while not done:
            confirmation = input("Would you like to continue? Y/N? ").lower()
            co()
            
            if confirmation == 'n':
                done = True
            elif confirmation == 'y':
                print('Thank you for playing!')
                game_over = False
                while not game_over:
                    print('Dealer Hand: ' + str(dealer.hand[0]) + ' + Mystery Card')
                    print(f'Total:  {Player.get_total(dealer)}')
                    print('Player Hand: ' + str(human.hand[0]) + ' ' + str(human.hand[1]))
                    print(f'Total:  {Player.get_total(human)} \n')
                    if Player.get_total(human) > 21:
                        co()
                        print('Player Hand: ' + str(human.hand))
                        print('You lose')
                        clear_hand()
                        break
                    choice = input("Would you like to hit or stand? ").lower()
                    if choice == 'hit':
                        dealer.deal(deck, human)
                        Player.get_total(human)
                    elif choice == 'stand':
                        while Player.get_total(dealer) < 17: 
                            dealer.deal(deck, dealer)
                        if Player.get_total(dealer) > 21: 
                            co()
                            print('Dealer ended with ' + str(Player.get_total(dealer)) +'\nDealer BUSTED')
                            clear_hand()
                            break
                        elif Player.get_total(dealer) > Player.get_total(human):
                            co()
                            print('Dealer ended with ' + str(Player.get_total(dealer)) + '\nDealer Wins...')
                            clear_hand()
                            break
                        elif Player.get_total(dealer) == Player.get_total(human):
                            co()
                            print('Dealer ended with ' + str(Player.get_total(dealer)) +'\nYou TIED')
                            clear_hand()
                            break
                        else: 
                            co()
                            print('Dealer ended with ' + str(Player.get_total(dealer)) +'\nYou WIN!!!!')  
                            clear_hand()
                            break
                        game_over = True
                        # Dealer starts dealing to themselves
                        # if dealer's hand has higher total rank than human, dealer wins
                            # Dealer wins
                        # if dealer's hand has total rank value higher than 21, dealer loses
                            # Dealer bust
                        # if dealer's hand has total rank value lower than the human's, dealer loses
                            # Dealer lose
        print('Thank you for playing!')
        



Game.run()
```
